```python
# Convert keywords to lower
word_list = keywords['Rewarding']
for i in range(len(word_list)):
  word_list[i] = word_list[i].lower()

print(keywords['Rewarding'])
```

```python
with open('Data/keywords_Incentives.json', 'w') as dict:
    json.dump(keywords, dict)
```

```python
#Keyword filter on Peru dataset

import pandas as pd
df = pd.read_csv('Diario_Oficial_Peru.csv')
df.head()
```

```python
title_incentivo = []
for title in df['title']:
  try:
    lower_list = title.lower()
    lower_list = lower_list.split(' ')
    for token in lower_list:
      if token in keywords['Rewarding']:  
        title_incentivo.append(title)
  except:
    #print('This title is not a string:', title)
    pass
```
