```python
# Download sample dataset
import pandas as pd
df = pd.read_csv('Diario_Oficial_Peru.csv')
df.head()
```

```python
title_ambiental = []
keywords_save = []
url_save = []
keyword_arr = numpy.array(list(keywords[topic] for topic in keywords))
for line in range(len(keyword_arr)):
  keyword_arr[line] = numpy.array(keyword_arr[line])
print(keyword_arr.shape)
print(keyword_arr[2])
title_arr = numpy.array(df['title'])
url_arr = numpy.array(df['url'])
for count, title in enumerate(title_arr):
  try:
    title_lower = title.lower()
    for arr in keyword_arr:
      for token in arr:
        if token in title_lower:  
          title_ambiental.append(title)
          keywords_save.append(token)
          url_save.append(url_arr[count])
  except:
    #print('This title is not a string:', title)
    pass

print(len(title_ambiental), len(keywords_save), len(url_save))  

title_pr = title_ambiental
keywords_pr = keywords_save
urls_pr = url_save
```

```python
# Write to CSV
import csv

with open("Peru_urls.csv",'w') as result_file:
  fieldnames = ['Title', 'Keywords', 'URL']
  writer = csv.DictWriter(result_file, fieldnames=fieldnames)
  writer.writeheader()
  for index in range(len(title_pr)):
    writer.writerow({'Title': title_pr[index], 'Keywords': keywords_pr[index], 'URL': urls_pr[index]})
```
